#### Content Based Filtering Using the Anime Dataset

```python
import pandas as pd
import numpy as np
```

```python
def recommend_anime(animeid, anime_df, num_recommendations=5):
    if animeid not in anime_df['anime_id'].values:
        return f"Anime ID {animeid} not found."

    input_anime = anime_df[anime_df['anime_id'] == animeid].iloc[0]
    print("Input Anime: ")
    display(pd.DataFrame([input_anime]))
    input_genres = set(input_anime['genre'].split(', '))

    similar_anime = anime_df[anime_df['anime_id'] != animeid].copy()
    
    def shared_genre_count(genres):
        if isinstance(genres, str):
            return len(input_genres.intersection(set(genres.split(', '))))
        return 0
    
    similar_anime['shared_genres'] = similar_anime['genre'].apply(shared_genre_count)

    recommended_anime = similar_anime.sort_values(by=['shared_genres', 'rating'], ascending=[False, False]).head(num_recommendations)
    return recommended_anime.drop('shared_genres',axis=1)


anime_df=pd.read_csv('./anime.csv')
animeid=int(input("Anime ID: "))
recommended=recommend_anime(animeid,anime_df,10)
print("Recommendations: ")
display(recommended)
```
